## Parte 2 — Variables: ponerle nombre a los números

Una variable es simplemente una **etiqueta** que pegás sobre un valor para poder reutilizarlo. En lugar de escribir `150` en todas partes, le decís que `precio = 150`.

El signo `=` en Python no significa *"igual"* matemático — significa *"guardá este valor con este nombre"*.

```python
# Definimos variables con datos de supply chain
demanda_anual   = 12000    # unidades por año
costo_pedido    = 250      # USD por orden emitida
costo_mantener  = 3        # USD por unidad almacenada por año
precio_unitario = 45       # USD por unidad

print("Demanda anual:", demanda_anual)
print("Costo de emitir una orden:", costo_pedido, "USD")
```

### Aplicación: Cantidad Económica de Pedido (EOQ)

Ya tenemos los datos guardados en variables. Ahora usamos la fórmula EOQ que conocés del módulo de inventarios:

$$EOQ = \sqrt{\dfrac{2 \times D \times S}{H}}$$

donde D = demanda anual, S = costo por orden, H = costo de mantenimiento.

```python
# Calculamos el EOQ usando las variables definidas arriba
# ** 0.5 es la raíz cuadrada
eoq = (2 * demanda_anual * costo_pedido / costo_mantener) ** 0.5

print(f"EOQ = {eoq:.0f} unidades")
print(f"Eso implica {demanda_anual / eoq:.1f} órdenes por año")
```

> 📌 **Nota sobre `f"..."`:** Es un *f-string*, la forma de Python para mezclar texto y variables en el mismo `print`. La `{variable}` se reemplaza por el valor. El `:.0f` significa *"mostrá sin decimales"*.

**Probalo vos:** Cambiá `demanda_anual` a 20000 y volvé a correr las dos últimas celdas. El EOQ se recalcula solo.

## Parte 3 — Listas: una columna de Excel

Una **lista** es una secuencia ordenada de valores. Pensala como una columna de Excel entre corchetes `[ ]`.

Con `sum()`, `len()`, `max()` y `min()` hacés las operaciones más comunes sin escribir nada complejo.

```python
# Demanda mensual de un producto (12 meses)
demanda_mensual = [980, 1050, 1200, 1350, 1400, 1550, 1600, 1480, 1300, 1100, 950, 800]

print("Demanda total anual: ", sum(demanda_mensual))
print("Demanda promedio:    ", sum(demanda_mensual) / len(demanda_mensual))
print("Mes pico:            ", max(demanda_mensual))
print("Mes valle:           ", min(demanda_mensual))
print("Cantidad de meses:   ", len(demanda_mensual))
```

```python
# Acceder a un elemento por posición
# ⚠️ En Python la posición empieza en 0, no en 1
print("Demanda de enero (posición 0):", demanda_mensual[0])
print("Demanda de julio (posición 6):", demanda_mensual[6])
print("Demanda de diciembre (último):", demanda_mensual[-1])  # -1 es el último
```

## Parte 4 — `if`: tomar decisiones

El `if` le dice a Python *"hacé esto solo si se cumple esta condición"*. Es la misma lógica que un `SI()` en Excel.

La **indentación** (el espacio al inicio de la línea) es obligatoria — así Python sabe qué está "dentro" del if.

```python
stock_actual    = 180   # unidades en depósito
punto_reorden   = 200   # umbral para emitir orden

if stock_actual < punto_reorden:
    print("⚠️  Stock bajo — emitir orden de compra")
else:
    print("✅  Stock suficiente — no se requiere acción")
```

**Probalo vos:** Cambiá `stock_actual` a 250 y corré de nuevo. Después cambialo a 100.

## Parte 5 — `for`: repetir para muchos SKUs

El `for` ejecuta el mismo bloque de código para cada elemento de una lista. Es la automatización más básica: en vez de revisar cada SKU a mano, Python lo hace por vos.

Un **diccionario** (`dict`) asocia una clave con un valor — como una tabla de dos columnas. Llave → Valor.

```python
# Diccionario: SKU → stock actual
inventario = {
    "SKU-A100": 350,
    "SKU-B220": 80,
    "SKU-C301": 12,
    "SKU-D450": 220,
    "SKU-E015": 5
}

punto_reorden = 100  # mismo umbral para todos

print("--- Revisión de inventario ---")
for sku, stock in inventario.items():
    if stock < punto_reorden:
        print(f"⚠️  {sku}: solo {stock} unidades — REPONER")
    else:
        print(f"   {sku}: {stock} unidades — OK")
```

**¿Qué hizo ese código?**  
- Tomó cada par `(sku, stock)` del diccionario.  
- Para cada uno, verificó si el stock estaba por debajo del umbral.  
- Imprimió un alerta o confirmación.

En Excel harías esto con `SI()` en cada fila. Acá, con 5 líneas, funciona para 5 SKUs o para 50,000 — sin cambiar nada.

## Parte 6 — pandas: la tabla de Excel de Python

**pandas** es la librería más usada para análisis de datos. Su objeto central es el `DataFrame` — básicamente, una tabla con columnas y filas, igual que una hoja de Excel.

La línea `import pandas as pd` carga la librería y le da el alias `pd` (por convención).

```python
import pandas as pd

# Creamos una tabla de órdenes de compra directamente en Python
datos = {
    "sku"          : ["SKU-A", "SKU-B", "SKU-C", "SKU-A", "SKU-B", "SKU-C"],
    "mes"          : [    "Ene",  "Ene",  "Ene",  "Feb",  "Feb",  "Feb"],
    "unidades"     : [    1200,    800,    450,   1350,    750,    500],
    "costo_unitario": [    12.5,   18.0,   25.0,   12.5,   18.0,   25.0]
}

df = pd.DataFrame(datos)
df  # escribir solo el nombre del DataFrame lo muestra como tabla
```

```python
# Agregar una columna calculada (como una columna fórmula en Excel)
df["costo_total"] = df["unidades"] * df["costo_unitario"]
df
```

```python
# Agrupar por SKU y sumar unidades (equivalente a una tabla dinámica)
resumen = df.groupby("sku")["unidades"].sum().reset_index()
resumen.columns = ["SKU", "Unidades Totales"]
resumen = resumen.sort_values("Unidades Totales", ascending=False)
resumen
```

```python
# Estadísticas descriptivas de una columna (como la función DESCRIPCIÓN de Excel)
df["unidades"].describe()
```

## Parte 7 — Un gráfico en una línea

Matplotlib es la librería de gráficos. En general no necesitás saber mucho — pandas tiene integración directa y con `.plot()` ya tenés un gráfico.

```python
import matplotlib.pyplot as plt

meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun",
         "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]

demanda_mensual = [980, 1050, 1200, 1350, 1400, 1550,
                   1600, 1480, 1300, 1100, 950, 800]

plt.figure(figsize=(10, 4))
plt.plot(meses, demanda_mensual, marker="o", color="steelblue", linewidth=2)
plt.axhline(sum(demanda_mensual)/12, color="red", linestyle="--", label="Promedio")
plt.title("Demanda mensual — SKU-A100")
plt.ylabel("Unidades")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
